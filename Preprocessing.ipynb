```python
df['preffered_premium_plan'] = df['preffered_premium_plan'].str.replace('-Rs 179/month', '')
df['preffered_premium_plan'] = df['preffered_premium_plan'].str.replace('- Rs 119/ month', '')
df['preffered_premium_plan'] = df['preffered_premium_plan'].str.replace('-Rs 59/month', '')
df['preffered_premium_plan'] = df['preffered_premium_plan'].str.replace('- Rs 149/month', '')
```

```python
df['spotify_subscription_plan'] = df['spotify_subscription_plan'].str.replace('(ad-supported)', '')
df['spotify_subscription_plan'] = df['spotify_subscription_plan'].str.replace('(paid subscription)', '')
```

```python
df.head()
```

```python
df[['Smartphone', 'Computer_or_laptop', 'Wearable_devices','Smart_speakers_or_voice_assistants']] = df['spotify_listening_device'].str.split(', ', expand=True)

# Display the DataFrame with split columns
df
```

```python
def split_based_on_device(row):
    devices = row['spotify_listening_device'].split(', ')
    return [device in devices for device in ['Smartphone', 'Computer or laptop', 'Smart speakers or voice assistants','Wearable_devices']]

# Apply the function to create new columns
listening_device_cols = ['Smartphone', 'Computer/laptop', 'Smart_speakers/voice_assistants','Wearable_devices']
df[listening_device_cols] = pd.DataFrame(df.apply(split_based_on_device, axis=1).tolist(), index=df.index)

# Drop the original column if needed
df.drop('spotify_listening_device', axis=1, inplace=True)
df.drop('Smart_speakers_or_voice_assistants', axis=1, inplace=True)
df.drop('Computer_or_laptop', axis=1, inplace=True)
df.head()
```

```python
sns.countplot(x='Gender',data=df,palette='viridis',hue='Age')
plt.show()
```

 Most of the spotify users lie between 20-35 years of age and most of them are female.
